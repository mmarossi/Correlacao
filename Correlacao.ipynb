```python
nome_do_arquivo="Estudo_Corr.xlsx"
df=pd.read_excel(nome_do_arquivo)
df.head(3)
```

```python
df.info()
```

```python
df.describe()
```

```python
df_hiper = df[df['Formato']=='HIPER']
df_super = df[df['Formato']=='SUPER']
```

```python
#gráfico de turnover por Regional Hiper
g1= sns.barplot(y='Regional', x='Turnover', data=df_hiper, palette="Blues")
g1.set_title("Turnover por Regional")
```

```python
df_hiper.groupby('Regional')['Turnover'].mean().sort_values(ascending=False)
```

```python
#gráfico de turnover por Regional Super
g2= sns.barplot(y='Regional', x='Turnover', data=df_super, palette="Blues")
g2.set_title("Turnover por Regional")
```

```python
df_super.groupby('Regional')['Turnover'].mean().sort_values(ascending=False)
```

```python
#gráfico de ABS por Regional Hiper
g3=sns.barplot(y='Regional', x='ABS', data=df_hiper, palette="Blues")
g3.set_title("Absenteísmo por Regional")
```

```python
df_hiper.groupby('Regional')['ABS'].mean().sort_values(ascending=False)
```

```python
#gráfico de ABS por Regional Super
g4=sns.barplot(y='Regional', x='ABS', data=df_super, palette="Blues")
g4.set_title("Absenteísmo por Regional")
```

```python
df_super.groupby('Regional')['ABS'].mean().sort_values(ascending=False)
```

```python
#gráfico de banco de horas por Regional Hiper
g5=sns.barplot(y='Regional', x='BancoHoras', data=df_hiper, palette="Blues")
g5.set_title("Banco de Horas por Regional")
```

```python
df_hiper.groupby('Regional')['BancoHoras'].mean().sort_values(ascending=False)
```

```python
#gráfico de banco de horas por Regional Super
g6=sns.barplot(y='Regional', x='BancoHoras', data=df_super, palette="Blues")
g6.set_title("Banco de Horas por Regional")
```

```python
df_super.groupby('Regional')['BancoHoras'].mean().sort_values(ascending=False)
```

```python
df1 = df_hiper[['Vendas(Mil)','Eficiencia', 'Perdas', 'BancoHoras', 'Turnover', 'ABS', 'NPS']]
correlation = df1.corr()
plot = sns.heatmap(correlation, annot = True, fmt=".1f", linewidths=0.6)
plot
```
